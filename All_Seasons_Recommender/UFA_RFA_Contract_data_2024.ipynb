## CONTRACT DETAILS - UFA AND RFA
### UFAs 2024:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# This is a hypothetical URL extracted from the iframe src attribute
url = 'https://puckpedia.com/e/players?q=3CKXYY&sz=750'

# Send an HTTP request and get the response
response = requests.get(url)

# Parse the response content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table in the parsed HTML
table = soup.find('table')  # You might need to refine this to find the exact table

# Convert the table to a pandas DataFrame
df_UFA_2024 = pd.read_html(str(table), index_col=0)[0]
df_UFA_2024
```

```python
#Dropping Goalies:
df_UFA_2024 = df_UFA_2024.drop(df_UFA_2024[df_UFA_2024.Pos == 'G'].index) #Removed Goalies sincenot considering them

#Changing the Column with the Player Name to be more appropriately named
df_UFA_2024.rename(columns={'Playeraverages': 'Player'}, inplace=True)
# Ensure that 'Player' column is of string type and not float or NaN
df_UFA_2024['Player'] = df_UFA_2024['Player'].astype(str)
# Split the string on comma, reverse, and join with a space
df_UFA_2024['Player'] = df_UFA_2024['Player'].apply(lambda x: ' '.join(x.split(',')[::-1]).strip())
df_UFA_2024['Player'] = df_UFA_2024['Player'].astype(str) # convert it back to a string
```

```python
# Drop Columns that have all NaNs or very few bits of data or game-related info since I have the players' statistics:
UFA_drop_cols = ['Team', 'Clause', 'Age', 'Height', 'Weight',  'Shoots','Pos','2023-24 GP', '2023-24 G', '2023-24 A',
                 '2023-24 P', '2023-24 PPG','2023-24 GPG', '2023-24 5x5 TOI', '2023-24 5x5 TOI/G',
                 '2023-24 GAA', '2023-24 SV%','2022-23 GF%', '2022-23 DFF%', '2022-23 relDFF%',
                 '2022-23 CF%', '2022-23 relCF%', '2022-23 G60', '2022-23 P60', '2022-23 GSAx', 'Signing Status', 'Start Yr']
```

```python
df_UFA_2024_dropped = df_UFA_2024.drop(columns=UFA_drop_cols)
```

```python
df_UFA_2024_dropped.info()
```

```python
df_UFA_2024_dropped.loc[df_UFA_2024_dropped['Agent'].isna(), 'Agent'] = 'No Agent'
```

```python
df_UFA_2024 = df_UFA_2024_dropped.copy()
```

### RFAs 2024:

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Replace with the actual URL you get from the iframe
url = 'https://puckpedia.com/e/players?q=33VEJC&sz=750'

# This will get the content from the URL, assuming it's not loaded dynamically
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Assuming the table is properly formatted and identifiable
table = soup.find('table')

# Now we convert it into a DataFrame
# index_col=0 will use the first column as the index
df_RFA_2024 = pd.read_html(str(table), index_col=0)[0]
```

```python
#Dropping Goalies:
df_RFA_2024 = df_RFA_2024.drop(df_RFA_2024[df_RFA_2024.Pos == 'G'].index) #Removed Goalies sincenot considering them

#Changing the Column with the Player Name to be more appropriately named
df_RFA_2024.rename(columns={'Playeraverages': 'Player'}, inplace=True)
# Ensure that 'Player' column is of string type and not float or NaN
df_RFA_2024['Player'] = df_RFA_2024['Player'].astype(str)
# Split the string on comma, reverse, and join with a space
df_RFA_2024['Player'] = df_RFA_2024['Player'].apply(lambda x: ' '.join(x.split(',')[::-1]).strip())
df_RFA_2024['Player'] = df_RFA_2024['Player'].astype(str) # convert it back to a string
```

```python
df_RFA_2024.loc[df_RFA_2024['Agent'].isna(), 'Agent'] = 'No Agent'
```

```python
# Drop Columns that have all NaNs or very few bits of data or game-related info since I have the players' statistics:
RFA_drop_cols = ['Team', 'Age', 'Height',  'Weight', 'Pos','Clause', '2023-24 GP', '2023-24 G', '2023-24 A',
                 '2023-24 P', '2023-24 PPG','2023-24 GPG', '2023-24 5x5 TOI', '2023-24 5x5 TOI/G',
                 '2023-24 GAA', '2023-24 SV%','2022-23 GF%', '2022-23 DFF%', '2022-23 relDFF%',
                 '2022-23 CF%', '2022-23 relCF%', '2022-23 G60', '2022-23 P60', '2022-23 GSAx', 'Shoots', 'Signing Status', 'Start Yr']
```

```python
set(df_RFA_2024_dropped.columns).intersection(set(df_UFA_2024_dropped.columns))
```

```python
frames = [df_UFA_2024_dropped, df_RFA_2024_dropped]

df_2024_NHL_expr_contract_details = pd.concat(frames, axis=0, join= 'outer',ignore_index=False)

df_2024_NHL_expr_contract_details.shape
```

```python
df_2024_NHL_expr_contract_details  = df_2024_NHL_expr_contract_details.reset_index(drop=True)
df_2024_NHL_expr_contract_details
```

```python
df_2024_NHL_expr_contract_details.to_csv('NHL_expr_contract_details_2024.csv', index=False)
```

## Bring in the All_strengths CSV so that I can use the team column from that one and create a team column in this one

```python
df_all_stats_AS_clean = pd.read_csv('NHL_Skater_AS_All_Stats_Clean.csv')
```

```python
# Make a df with just the 2024 stats so I can link the team to the contract table:
df_AS_2024 = df_all_stats_AS_clean.loc[df_all_stats_AS_clean['Season'] == 2024]
df_AS_2024
```

```python
df_merged = pd.merge(df_2024_NHL_expr_contract_details,
                     df_AS_2024[['Player', 'Team']],
                     on='Player', how='left')

df_merged.loc[df_merged['Team'].isna()]
```

```python
df_merged_alt = pd.merge(df_AS_2024[['Player', 'Team']],
                         df_2024_NHL_expr_contract_details,
                         on='Player', how='left')

df_merged_alt_missing = df_merged_alt.loc[df_merged_alt['Current  Cap Hit'].isna()]

df_exp_contracts_2024 = df_merged_alt.drop(df_merged_alt_missing.index)
df_exp_contracts_2024.reset_index(drop=True)
```

```python
df_exp_contracts_2024.to_csv('NHL_expr_contract_details_2024.csv', index=False)
```
